# Q5- What are iterators in python and how do they differ from iterables?

Iterables
Definition: An iterable is any Python object capable of returning its members one at a time. It is an object that can be looped over (iterated over) using a for loop or comprehended using a comprehension. Common examples include lists, tuples, strings, and dictionaries.

Characteristics:

Definition: Iterables implement the __iter__() method, which returns an iterator object. Alternatively, they implement the __getitem__() method, which allows indexed access.
Usage: You can iterate over an iterable using a for loop or pass it to functions like list(), set(), or sum().

Example:

```python
# List is an iterable
my_list = [1, 2, 3, 4, 5]

# You can iterate over the list
for item in my_list:
    print(item)
```

Iterators
Definition: An iterator is an object that represents a stream of data. It produces items one at a time and maintains its state as it iterates. An iterator is an object that follows the iterator protocol: it implements the __iter__() and __next__() methods.

Characteristics:

Definition: Iterators implement the __iter__() method (which returns the iterator object itself) and the __next__() method (which returns the next item in the sequence). When there are no more items to return, __next__() raises the StopIteration exception to signal the end of iteration.
Usage: You can use an iterator to traverse through all elements of an iterable one at a time, usually in a for loop or with the next() function.

Example:

```python
# Create an iterator from a list
my_list = [1, 2, 3, 4, 5]
iterator = iter(my_list)  # Get an iterator from the iterable

# Manually iterate over the iterator
print(next(iterator))  # Outputs: 1
print(next(iterator))  # Outputs: 2
# and so on...
```

# Q6- Explain the concepts of generators in python and how they are defined.


A generator is a special type of iterator that is defined using a function but yields values one at a time, using the yield keyword. Generators are used to produce a sequence of values over time, rather than computing all the values upfront and storing them in memory. This can be particularly useful for working with large datasets or streams of data.

Key Characteristics of Generators
Lazy Evaluation:

Generators produce items one at a time and only when required, which means they are more memory-efficient than using lists or other collections that require holding all values in memory.
Stateful Iteration:

Generators maintain their state between iterations. They remember where they left off and resume from that point when next() is called.
Simplified Code:

They allow you to write code that is more readable and concise compared to manually implementing an iterator class.
Defining Generators
Generators are defined using functions with the yield keyword. Unlike functions that return a single value and exit, generator functions can yield multiple values, pausing and resuming their execution.

Example of a Generator Function:

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

# Using the generator
counter = count_up_to(5)
for num in counter:
    print(num)  # Outputs: 1, 2, 3, 4, 5
```

# Q7- What are the advantages of using generators over regular functions?

1. Memory Efficiency
Advantage: Generators are memory-efficient because they generate values one at a time and do not require the entire sequence to be stored in memory.

Regular Functions: Typically return an entire list or collection, which can be memory-intensive for large datasets.
Generators: Yield values on-the-fly, which means only one value is stored in memory at any given time.

Example:

```python
# Regular function (memory-intensive)
def generate_large_list(n):
    return [i for i in range(n)]

large_list = generate_large_list(10**6)  # Stores a large list in memory

# Generator (memory-efficient)
def generate_large_numbers(n):
    for i in range(n):
        yield i

large_generator = generate_large_numbers(10**6)  # Does not store values, generates on demand
```

2. Performance Improvement
Advantage: Generators can improve performance by producing values only when needed, reducing the initial computation time and avoiding the overhead of creating and storing large data structures.

Regular Functions: Compute and return the entire result set in one go, which may be slow for large datasets.
Generators: Compute values as needed, which can be faster and allows for more responsive and interactive applications.

Example:

```python
# Using a generator
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

# Efficiently iterates over values without computing all at once
for num in count_up_to(10**6):
    pass  # Process numbers one by one
```

3. Infinite Sequences
Advantage: Generators can be used to create infinite sequences, something that is not feasible with regular functions that generate finite collections.

Regular Functions: Cannot return infinite sequences because they require allocating memory for all elements.
Generators: Can handle infinite sequences since they generate values on-the-fly.

Example:

```python
# Generator for an infinite sequence
def infinite_count(start=0):
    while True:
        yield start
        start += 1

for num in infinite_count():
    if num > 10:
        break  # Stop after a certain condition
    print(num)
```

4. Lazy Evaluation
Advantage: Generators use lazy evaluation, meaning they produce values only when required. This can lead to significant performance benefits when dealing with large datasets or streams of data.

Regular Functions: Evaluate all values at once, even if only a subset is needed.
Generators: Evaluate and yield values as needed, saving computation and memory.

Example:

```python
# Generator that reads lines from a file
def read_lines(file_name):
    with open(file_name, 'r') as file:
        for line in file:
            yield line

# Reads file line by line, only loading one line into memory at a time
for line in read_lines('large_file.txt'):
    process(line)  # Process each line as it is read
```

5. Simplified Code
Advantage: Generators can simplify code by eliminating the need for explicit iterator classes and maintaining state between iterations.

Regular Functions: May require additional code to manage state and iteration logic.
Generators: Use yield to manage iteration and state concisely within a single function.

Example:

```python
# Regular function with explicit iterator class
class Counter:
    def __init__(self, low, high):
        self.current = low
        self.high = high

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.high:
            raise StopIteration
        self.current += 1
        return self.current - 1

# Simplified with a generator
def counter(low, high):
    while low <= high:
        yield low
        low += 1
```

Return the total sum of even numbers.
Hereâ€™s a Python function that accomplishes this:

```python
def sum_of_even_numbers(numbers):
    # Initialize the sum variable
    total_sum = 0

    # Iterate through the list
    for number in numbers:
        # Check if the number is even
        if number % 2 == 0:
            # Add the even number to the total sum
            total_sum += number

    # Return the total sum of even numbers
    return total_sum

# Example usage
numbers = [47, 11, 42, 13]
print(sum_of_even_numbers(numbers))  # Output: 42
```

# Q4- Write a Python function that checks if a given number is prime or not from 1 to 200.

To write a Python function that checks if a given number is prime and also ensure that it only works for numbers between 1 and 200, you need to:

Check if the number is within the valid range (1 to 200).
Determine if the number is a prime number.
A prime number is a number greater than 1 that has no positive divisors other than 1 and itself.

Hereâ€™s how you can implement such a function:

```python
def is_prime(number):
    # Check if the number is within the valid range
    if number < 1 or number > 200:
        return False

    # 1 and numbers less than 1 are not prime
    if number <= 1:
        return False

    # Check for factors from 2 up to the square root of the number
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False

    return True

# Example usage
test_numbers = [1, 2, 3, 4, 5, 200, 199, 101]
results = {num: is_prime(num) for num in test_numbers}
print(results)
```

Explanation
Range Check: The function first checks if the number is within the range [1, 200]. If not, it returns False.

Prime Check:

Numbers less than or equal to 1 are not prime.
For numbers greater than 1, the function checks divisibility from 2 up to the square root of the number (rounded down) using int(number**0.5) + 1. This is because a larger factor of the number would have to be paired with a smaller factor that would have already been checked.
If any number divides number evenly (i.e., number % i == 0), then number is not prime.
Return Result: If no divisors are found, the number is prime, and the function returns True.

# Q5- Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

To create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms, you'll need to define a class with the necessary methods to support iteration. The key components are:

Initialization: The class should accept the number of terms to generate.
Iterator Methods: Implement __iter__() and __next__() methods to make the class an iterator.

Hereâ€™s a step-by-step implementation:


```python
class FibonacciIterator:
    def __init__(self, n):
        """Initialize with the number of terms."""
        self.n = n
        self.current = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        """Return the iterator object itself."""
        return self

    def __next__(self):
        """Return the next value in the Fibonacci sequence."""
        if self.current < self.n:
            # The current value in the sequence
            result = self.a
            # Update Fibonacci numbers
            self.a, self.b = self.b, self.a + self.b
            self.current += 1
            return result
        else:
            # Stop iteration if n terms have been generated
            raise StopIteration

# Example usage
fibonacci_sequence = FibonacciIterator(10)
for number in fibonacci_sequence:
    print(number)
```

Explanation
Initialization (__init__):

self.n: Stores the total number of terms to generate.
self.current: Keeps track of the number of terms generated so far.
self.a and self.b: Initialize the first two Fibonacci numbers (0 and 1).
Iterator Method (__iter__):

Returns the iterator object itself. This is required for the iterator protocol.
Next Value Method (__next__):

Checks if the number of terms generated (self.current) is less than the specified number (self.n).
If true, returns the current Fibonacci number (self.a) and updates self.a and self.b to the next values in the sequence.
If the number of terms generated is equal to or greater than self.n, it raises StopIteration to signal that the iteration is complete.

# Q6- Write a generator function in Python that yields the powers of 2 up to a given exponent.

Certainly! A generator function in Python provides a way to iterate over a sequence of values without generating the entire sequence at once. To create a generator that yields the powers of 2 up to a given exponent, you can use the yield keyword.

Hereâ€™s how you can write such a generator function:

```python
def powers_of_two(max_exponent):
    """Generator that yields powers of 2 up to 2^max_exponent."""
    exponent = 0
    while exponent <= max_exponent:
        yield 2 ** exponent
        exponent += 1

# Example usage
for power in powers_of_two(5):
    print(power)
```

This function, powers_of_two, takes a max_exponent as an argument and yields 2 raised to each exponent from 0 to max_exponent. You can call it and iterate through the results to see the powers of 2.

#Q10- Create a Python program that uses `filter()` to remove all the vowels from a given string.

You can create a Python program that uses the filter() function to remove all vowels from a given string. Here's a simple implementation:

```python
def remove_vowels(input_string):
    # Define a set of vowels
    vowels = set('aeiouAEIOU')

    # Use filter to remove vowels
    filtered_string = ''.join(filter(lambda char: char not in vowels, input_string))

    return filtered_string

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)
print(result)  # Output: Hll, Wrld!
```

Explanation:
Vowels Set: A set of vowels (both lowercase and uppercase) is defined.
Filter Function: The filter() function is used with a lambda function that checks if each character is not in the vowels set.
Join: The resulting filtered characters are joined back into a string using join().
